```python
import sys
sys.path.append('../')

import pandas as pd
import jsonpickle
from measure.metrics import *
from measure.tasks import *

NRMSE = lambda y,y_fit: np.mean(((y-y_fit)**2)/np.var(y))
```

```python
fitness = pd.read_parquet('../fitness.parquet')
fitness = fitness.sort_values('fitness', ascending=True)  
lowest_fitness_row = fitness.iloc[0]
encoded_final_res = lowest_fitness_row['final_reservoir']
res_lowest_fitness = jsonpickle.decode(encoded_final_res)
print('fitness:', lowest_fitness_row['fitness'])
KR, GM = get_metrics(res_lowest_fitness.bipolar())

input, target = narmax(order=10)
res_lowest_fitness.reset()
predictions = res_lowest_fitness.bipolar().train(input, target)
err = np.min((NRMSE(target[:, res_lowest_fitness.washout:], predictions), 1))

print('narma-10:', err)
print('KR:', KR)
print('GM:', GM)
# print('MC:', linear_memory_capacity(res=final_graph_with_lowest_fitness.bipolar()))
print('num_nodes:', res_lowest_fitness.size())
res_lowest_fitness.bipolar().draw_gt()
```
